```python
import os
from PIL import Image
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
folder_path = os.getcwd()
pristine_image_path = os.path.join(folder_path, 'phase-01-training', 'dataset-dist', 'phase-01', 'training', 'pristine')
fake_image_path = os.path.join(folder_path, 'phase-01-training', 'dataset-dist', 'phase-01', 'training', 'fake')
print("parent directory", folder_path)
print("pristine image directory", pristine_image_path)
print("fake image directory", fake_image_path)
```

```python
print("No of pristine images", len(os.listdir(os.path.join(folder_path, 'phase-01-training', 
                                            'dataset-dist', 'phase-01', 'training', 'pristine'))))
print("No of fake images", (len(os.listdir(os.path.join(folder_path, 'phase-01-training', 
                                            'dataset-dist', 'phase-01', 'training', 'fake')))-1)/2)
```
